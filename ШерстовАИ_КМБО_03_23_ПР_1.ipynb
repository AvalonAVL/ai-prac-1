# 4.3.2
Дана функция $\frac{\sqrt{1+e^\sqrt{x}+\cos{x^2}}}{|1-\sin^3{x}|}+\ln{|2x|};$ <br>
Создать массив из 10 значений функции ($x$, например, изменяется от 1 до 10) Выделить срез первой половины массива и построить графики для основного массива, используя ```plt.plot()```, для среза — точечный ```plt.scatter()```

````python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1, 11)
y = (np.sqrt(1 + np.exp(np.sqrt(x)) + np.cos(x ** 2))) / np.abs(
    1 - (np.sin(x)) ** 3) + np.log(2 * x)

plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
````

````python
half = len(x) // 2
plt.scatter(x[:half], y[:half])
plt.xlabel("x")
plt.ylabel("y")
plt.show()
````

# 4.3.3
Дана функция $|\cos{(xe^{(\cos{(x)}+\ln{(x+1)})})}|$. Построить график на интервале $(0,10)$ с шагом 1 с заливкой площади и найти эту площадь под ним. Для вычисления площади воспользуйтесь функцией ```trapz()```, вычисляющей интеграл по правилу трапеции. Для её корректной работы необходимо подключить следующие библиотеки: <br>
```from scipy.integrate import simps``` <br>
```from numpy import trapz```

````python
import numpy as np
import matplotlib.pyplot as plt
import math as m
from scipy.integrate import simps

x = np.arange(11)
y = np.abs(np.cos(x * np.exp(np.cos(x) + np.log(x + 1))))

area = np.trapz(y)
print(area)

plt.grid()
plt.plot(x, y, c = "b")
plt.fill_between(x, y)
plt.show()
````

# 4.3.4
Откройте статистические данные по стоимости акций Apple,
Microsoft, Google за 12 месяцев 2021 г. Постройте 3 графика на

Для начальных данных допустимо использовать следующий ресурс:

````python
from requests import get
from io import StringIO
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime


apple_url = "https://raw.githubusercontent.com/AvalonAVL/ai-prac-1/main/AAPL.csv"
microsoft_url = "https://raw.githubusercontent.com/AvalonAVL/ai-prac-1/main/MSFT.csv"
google_url = "https://raw.githubusercontent.com/AvalonAVL/ai-prac-1/main/GOOG.csv"
request = get(apple_url)
apple = pd.read_csv(StringIO(request.text))
request = get(microsoft_url)
microsoft = pd.read_csv(StringIO(request.text))
request = get(google_url)
google = pd.read_csv(StringIO(request.text))

dates = pd.date_range(start="2021-01-01",end="2021-12-31")

plt.grid()
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=31))
plt.plot(apple["Date"], apple["Close"], label = "Apple")
plt.plot(microsoft["Date"], microsoft["Close"], label = "Microsoft")
plt.plot(google["Date"], google["Close"], label = "Google")
plt.gcf().autofmt_xdate()
plt.legend()
plt.title("Stock prices")
plt.ylabel("Price, $")
plt.show()
````

# 4.3.5
Создайте простейший калькулятор, включающий основные действия для двух переменных $+, -, *, /$, а также вычисление следующих функций: $e^{x+y}, \sin{(x+y)}, \cos{(x+y)}, x^y$

````python
import math as m

def calculator():
  while True:
    x = float(input("Введите первый операнд: "))
    y = float(input("Введите второй операнд: "))
    print("Выберите действие (введите номер):", "1. +", "2. -", "3. *", "4. /", "5. e^(x+y)", "6. sin(x + y)",
          "7. cos(x + y)", "8. x**y", sep='\n')
    act = int(input())
    if act == 1:
      print(x, "+", y, "=", x + y)
    elif act == 2:
      print(x, "-", y, "=", x - y)
    elif act == 3:
      print(x, "*", y, "=", x * y)
    elif act == 4:
      print(x, "/", y, "=", x / y)
    elif act == 5:
      print("e^(", x, " + ", y, ") = ", m.exp(x + y), sep = "")
    elif act == 6:
      print("sin(", x, " + ", y, ") = ", m.sin(x + y), sep = "")
    elif act == 7:
      print("cos(", x, " + ", y, ") = ", m.cos(x + y), sep = "")
    elif act == 8:
      print(x, "**", y, "=", x ** y)
    else:
      print("Данная операция не поддерживается")

    print("Желаете подолжить? (y/n)")
    cont = input()
    if cont in ["n", "N", "no", "No", "н", "нет", "Нет"]:
      return
    print()

calculator()
````
